## Shuffling the dataset rows and creating train (4000 rows) and test (1000 rows) csv files 

```python
import csv
import random

path = "/kaggle/input/iitjee-neet-aims-students-questions-data/subjects-questions.csv"

train_file = open('train.csv', mode='w')
test_file = open('test.csv', mode='w')

train_writer = csv.writer(train_file, delimiter=',')
test_writer = csv.writer(test_file, delimiter=',')

with open(path, mode='r') as csv_file:
    csv_reader = csv.reader(csv_file)
    header = next(csv_reader) # skip header row
    train_writer.writerow(header) # write header row to train file
    test_writer.writerow(header) # write header row to test file
    
    rows = list(csv_reader)
    random.shuffle(rows) # shuffle rows randomly
    
    train_rows = rows[:4000] # first 4000 rows for training
    test_rows = rows[4000:5000] # next 1000 rows for testing
    
    # write rows to train and test files
    for row in train_rows:
        train_writer.writerow(row)
    for row in test_rows:
        test_writer.writerow(row)

train_file.close()
test_file.close()

```

```python
# Path to train and test
path_train = '/kaggle/working/train.csv'
path_test = '/kaggle/working/test.csv'

```

```python
# The example CHEMISTRY text to classify
text = """the branch of science that deals with the properties, composition, 
                          and structure of elements and compounds, how they can change, 
                          and the energy that is released or absorbed when they  change."""
```

## TextBlob Naive Bayes Classifier

```python
with open(path_train, 'r') as f:
    nb_cl = NaiveBayesClassifier(f, format='csv' )

with open(path_test, 'r') as f:
    print("Test of Accuracy Naive Bayes: ",nb_cl.accuracy(f, format='csv')) # 85 % accuracy
```

```python
print("This text belongs to: ", nb_cl.classify(text)) # Predicts correctly Chemistry
```

## Support Vector Machine (SVM) using Scikit-Learn

```python
# Load the CSV file
with open(path, 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    next(reader) # Skip the header row
    data = []
    labels = []
    for row in reader:
        data.append(row[0])
        labels.append(row[1])



# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.3)

# Create a feature extraction pipeline
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)

# Create an SVM classifier and fit on the training data
classifier = LinearSVC()
classifier.fit(X_train, y_train)


# Evaluate the performance of the classifier
accuracy = classifier.score(X_test, y_test)

print("Test of Accuracy of SVM:", accuracy) # 92 % accuracy

```

```python
# Use the classifier to predict the labels for the Text)

X_new = vectorizer.transform([text])
y_new = classifier.predict(X_new)
print("SVM: ",y_new) # Predicts correctly Chemistry
```
